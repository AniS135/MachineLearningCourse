```python
import cv2
import matplotlib.pyplot as plt
```

```python
b_l = cv2.imread('Dataset/bottom_left.jpg')
bottom_left = cv2.cvtColor(b_l, cv2.COLOR_BGR2RGB)
```

```python
print(bottom_left.shape)
```

```python
plt.imshow(bottom_left)
plt.show()
```

```python
bottom_left = cv2.resize(bottom_left, (200,200))
```

```python
plt.imshow(bottom_left)
plt.title("Image" + str(bottom_left.shape))
plt.show()
```

```python
b_r = cv2.imread('Dataset/bottom_right.jpg')
bottom_right = cv2.cvtColor(b_r, cv2.COLOR_BGR2RGB)
```

```python
print(bottom_right.shape)
```

```python
plt.imshow(bottom_right)
plt.show()
```

```python
bottom_right = cv2.resize(bottom_right, (200,200))
```

```python
plt.imshow(bottom_right)
plt.title("Image" + str(bottom_right.shape))
plt.show()
```

```python
t_l = cv2.imread('Dataset/top_left.jpg')
top_left = cv2.cvtColor(t_l, cv2.COLOR_BGR2RGB)
plt.imshow(top_left)
plt.show()
```

```python
top_left = cv2.resize(top_left, (200,200))
plt.imshow(top_left)
plt.title("Image" + str(top_left.shape))
plt.show()
```

```python
t_r = cv2.imread('Dataset/top_right.jpg')
top_right = cv2.cvtColor(t_r, cv2.COLOR_BGR2RGB)
plt.imshow(top_right)
plt.show()
```

```python
top_right = cv2.resize(top_right, (200,200))
plt.imshow(top_right)
plt.title("Image" + str(top_right.shape))
plt.show()
```

```python
import numpy as np
```

```python
col1 = np.vstack([top_left, bottom_left])
col2 = np.vstack([top_right, bottom_right])
collage = np.hstack([col1, col2])

collage = cv2.cvtColor(collage, cv2.COLOR_BGR2RGB)

plt.imshow(collage)
```
