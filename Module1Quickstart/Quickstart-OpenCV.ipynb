```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/Profile.jpeg')
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
print(img.shape)
```

```python
crop_image = rgb_image[60: , 40: , :]
plt.imshow(crop_image) # OpenCV read image in B G R format
# But we want to display it R G B for that
plt.show()
```

```python
crop_image.shape
```

```python
# Resize the image
```

```python
crop_image = cv2.resize(crop_image,(100, 100))
```

```python
plt.imshow(crop_image)
plt.title("Image" + str(crop_image.shape))
plt.show()
```
