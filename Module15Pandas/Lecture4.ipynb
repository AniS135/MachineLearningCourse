```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movie_metadata.csv")
```

```python
columns = df.columns
columns
```

```python
titles = list(df.get('movie_title'))
```

```python
dict = {}

for title in titles:
    if dict.get(len(title)) is None:
        dict[len(title)] = 1
    else:
        dict[len(title)] += 1
```

```python
X = []
for i in dict.keys():
    X.append((int(i)))
Y = []
for i in dict.values():
    Y.append((int(i)))

print(X)
print(Y)
```

```python
plt.scatter(x = X, y = Y)
plt.xlabel("Movies Length")
plt.ylabel("Frequency")
plt.title("Freq of Length of Movies Title")
plt.show()
```
