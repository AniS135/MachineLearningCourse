```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn')
```

```python
x_coordinates = np.arange(3) * 2
plt.bar(x_coordinates - 0.25, [10, 20, 15], width= 0.5, label = "Current Year", tick_label = ["Gold", "Platinum", "Silver"]) # Current Year
plt.bar(x_coordinates + 0.25, [20, 10, 12], width=0.5, label = "Next Year") # Next Year
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.style.use('dark_background')
x_coordinates = np.arange(3) * 2
plt.bar(x_coordinates - 0.25, [10, 20, 15], width= 0.5, label = "Current Year", tick_label = ["Gold", "Platinum", "Silver"], color = 'blue') # Current Year
plt.bar(x_coordinates + 0.25, [20, 10, 12], width=0.5, label = "Next Year", color = 'orange') # Next Year
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0, 30)
plt.legend()
plt.show()
```
