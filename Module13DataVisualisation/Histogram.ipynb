```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xsn = np.random.randn(100)
```

```python
sigma = 8
u = 70
x = np.round(xsn * sigma + u)
x2 = np.round(xsn * 5 + 40)
print(x, x2)
```

```python
plt.style.use('seaborn')
```

```python
plt.hist(x, alpha = 0.8, label="Maths") # Alpha -> Obacity constant
plt.hist(x2, alpha = 0.8, label="Physics")
plt.ylabel("Prob/Freq of Students")
plt.xlabel("Marks Range")
plt.legend()
plt.show()
```
