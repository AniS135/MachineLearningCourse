```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
prices = np.array([1, 2, 3, 4]) ** 3
print(prices)
```

```python
plt.plot(prices) # (x, y) if we give just one parameter then x is taken index automatically
plt.show()
```

```python
x = np.arange(10)
y1 = x ** 2
y2 = 2 * x + 3
```

```python
# Adjust the size of any plot
plt.figure(figsize=(4, 4))
plt.scatter(x, y1, color = 'red', label= "Apple", marker = 'o')
plt.scatter(x, y2, color = 'green', label= "Kiwi", linestyle = "dashed", marker = '*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruit over Time")
plt.show()
```
