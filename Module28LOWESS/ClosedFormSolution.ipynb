```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X, Y = make_regression(n_samples= 400, n_features= 1, noise=1.8, random_state= 42)

Y = Y.reshape((-1, 1))
print(X.shape)
print(Y.shape)
```

```python
X = (X - X.mean()) / X.std()
```

```python
plt.figure()
plt.scatter(X, Y)
plt.title("Normailzed Data")
plt.show()
```

```python
ones = np.ones((X.shape[0], 1))

X_ = np.hstack((X, ones))

print(X_.shape)
print(X_[:5, :])
```

```python
def predict(X, theta):
    return np.dot(X, theta)

def getThetaClosedForm(X, Y):

    Y = np.mat(Y)
    
    firstPart = np.dot(X.T, X)
    secondPart = np.dot(X.T, Y)

    theta = np.linalg.pinv(firstPart) * secondPart

    return theta
```

```python
theta = getThetaClosedForm(X_, Y)
print(theta)
```

```python
plt.figure()
plt.scatter(X, Y)
plt.plot(X, predict(X_, theta), color= "red")
plt.title("Normalize Data")
plt.show()
```
