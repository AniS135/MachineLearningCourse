```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read an Image

img = cv2.imread('Data/dog.png')

newImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# OpenCV by default read image as BGR
# So to get right image we have to switch from BGR to RGB
```

```python
plt.imshow(newImg)
plt.show()
```
