```python
import matplotlib.pyplot as plt
import numpy as np
```

### Multivariate Normal Distribution

```python
#BiVariate

mean = np.array([0.0, 0.0])
cov = np.array([[0.3, 0], [0, 1]])

mean2 = np.array([5.0, 6.0])
cov2 = np.array([[1.3, 0.2], [0.2, 1.1]])

dist = np.random.multivariate_normal(mean=mean, cov=cov, size=500)
dist2 = np.random.multivariate_normal(mean=mean2, cov=cov2, size=500)
print(dist.shape)

plt.scatter(dist[:, 0], dist[:, 1])
plt.scatter(dist2[:, 0], dist2[:, 1])
plt.show()
```
