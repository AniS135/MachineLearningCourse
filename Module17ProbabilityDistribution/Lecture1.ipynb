```python
import matplotlib.pyplot as plt
import numpy as np
```

### Histogram

```python
# Generate a Normal Distribution (1 - D)

u = 5
sigma = 0.2
# Such a distribution is also called standard normal distribution

vals = u + sigma * np.random.randn(200) # In randn, n stands for normal distribution

print(vals.shape)

plt.hist(vals, 10) # 2nd argument is number of buckets
# It denotes the number of values falling in interval
plt.show()
```

```python
x = vals
y = np.zeros(x.shape)

plt.scatter(x, y)
plt.show()
```
