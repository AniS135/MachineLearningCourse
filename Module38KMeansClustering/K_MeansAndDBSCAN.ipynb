### Problem with K-Means
- Can fail sometimes on complex datasets

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles, make_moons
```

```python
X, Y = make_moons(n_samples= 200, noise= 0.10)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = Y)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters= 2)
kmeans.fit(X)
```

```python
centers = kmeans.cluster_centers_
label = kmeans.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c = label)
plt.scatter(centers[:, 0], centers[:, 1], marker= "*", c = "red")
plt.show()
```

### DBSCAN
- Density Based Spatial Clustering of Applications with Noise
- Based upon on the idea, that a cluster is a high density area surrounded by low density region
- Starts by exploring a small area, if density is "good enough" it considered as part of cluster, and explores neighbours and so on, so it group regions based upn density 

```python
plt.scatter(X[:, 0], X[:, 1], c = Y)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps=0.2, min_samples= 5)
```

```python
dbs.fit(X)
```

```python
Ypred = dbs.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = Ypred)
plt.show()
```
