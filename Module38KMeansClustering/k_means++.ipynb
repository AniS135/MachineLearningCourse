### K-Means++
- K-Means is quite sensitive to initalization, if the init is not good, our algorithm is not able to make desired number os clusters.
- To overcome this problem, we use technique called K-Means++ which chooses initial centres so they are statiscally close to final ones.
- The method is default choice for sklearn, so we will use it directly without going into much maths! 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples = 500, centers= 5, random_state= 3)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters= 5)
```

```python
kmeans.fit(X)
```

```python
centers = kmeans.cluster_centers_
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(centers[:, 0], centers[:, 1], marker="*", color = "black")
plt.show()
```

```python
pred = kmeans.labels_
```

```python
print(pred)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = pred)
plt.scatter(centers[:, 0], centers[:, 1], marker="*", color = "black")
plt.title("Prediction")
plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.title("Actual")
plt.show()
```
