```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.arange(20)
print(X)
```

```python
Theta = np.array([2, 3])

noise = 3 * np.random.randn(20) # This function gives samples from normal distribution
```

```python
y_ideal = Theta[1] * X + Theta[0]
y_real = Theta[1] * X + Theta[0] + noise
```

```python
plt.plot(X, y_ideal, color = 'red')
plt.scatter(X, y_real)
plt.scatter(X, noise)
plt.show()
```
